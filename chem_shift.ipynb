```python
#get the starting index in pdb file
def get_starting_index(refer,refer_num,standard):
    index_0=int(refer_num[0])
    index_1=int(refer_num[1])
    index_2=int(refer_num[2])
    index_3=int(refer_num[3])
    if int(bmrb_num[0])+int(bmrb_num[3])==int(bmrb_num[1])+int(bmrb_num[2]):
        i=0
        while i<len(standard):
            if standard[i:i+4]!=refer[:4]:
                i+=1
            else:
                return i
    else:
        i=0
        while i<len(standard):
            if standard[i]==refer[0] and standard[i+index_1-index_0]==refer[1] and standard[i+index_2-index_0]==refer[2] and standard[i+index_3-index_0]==refer[3]:
                return i
            else:
                i+=1
    return 'NotFound'


```
